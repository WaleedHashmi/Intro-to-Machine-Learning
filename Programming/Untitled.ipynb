```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = []
x2 = []

for i in range (150):
    x1.append(X[i][0])
    x2.append(X[i][1])

      
plt.plot(x1,x2,'bo', color='green', marker='o', markersize=3)
plt.show()

```

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array(X)
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
pred = kmeans.labels_
```

```python
c0 = [[],[]]
c1 = [[],[]]
c2 = [[],[]]

for i in range(len(pred)):
    if pred[i] == 0:
        c0[0].append (X[i][0])
        c0[1].append (X[i][1])
    if pred[i] == 1:
        c1[0].append (X[i][0])
        c1[1].append (X[i][1])    
    if pred[i] == 2:
        c2[0].append (X[i][0])
        c2[1].append (X[i][1])
        
plt.plot(c0[0],c0[1], 'bo', color='green', marker='o', markersize=3)
plt.plot(c1[0],c1[1], 'bo', color='red', marker='o', markersize=3)
plt.plot(c2[0],c2[1], 'bo', color='blue', marker='o', markersize=3)



plt.show()
        
```
